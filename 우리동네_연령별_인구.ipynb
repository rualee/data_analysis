```python
f = open(file_name, encoding='UTF8') #open하면 메모리에 저장됨
data = csv.reader(f)
result = []

#우리동네 데이터만 확인
for row in data:
    if '서원동' in row[0]: #서원동 데이터만
        for i in row[3:]: # 0~100세 이상까지 모든 연령에 대해 반복하기
            #print(i)
            result.append(i) #해당 연령의 인구수 리스트에 순서대로 저장하기
print(result)
f.close() #close는 반드시!!!
```

```python
# X 축 타이틀 뽑기
f = open(file_name, encoding='UTF8')
data = csv.reader(f)
header = next(data)
#header

#연령만 글자만 추출
fields = []
for h in header[3: ]:
    #print(h.split('_'[-1])
    fields.append(h.split('_')[-1])
print(fields)

#우리동네 연령별 데이터
result = []

for row in data:
    if '서원동' in row[0]: #서원동 데이터만
        for i in row[3:]: # 0~100세 이상까지 모든 연령에 대해 반복하기
            #print(i)
            result.append(int(i.replace(",",""))) #해당 연령의 인구수 리스트에 순서대로 저장하기,콤마 지우고 문자-숫자로 형변환
print(result)
f.close()
```

# 최종 결과 코드

```python
def townPopulation(town):
    f = open(file_name, encoding='UTF8')
    data = csv.reader(f)
    #x축 타이틀 뽑기
    header = next(data)
    print(header)
#header

#연령만 글자만 추출
    fields = []
    for h in header[3: ]:
        #print(h.split('_'[-1])
        fields.append(h.split('_')[-1])
    print(fields)

    #우리동네 연령별 데이터
    result = []
    for row in data:
        if town in row[0]: #서원동 데이터만
            for i in row[3:]: # 0~100세 이상까지 모든 연령에 대해 반복하기
                #print(i)
                result.append(int(i.replace(",",""))) #해당 연령의 인구수 리스트에 순서대로 저장하기,콤마 지우고 문자-숫자로 형변환
        
    print(result)
    f.close()

    import matplotlib.pyplot as plt
    plt.style.use('ggplot') #격자 무늬 스타일 지정
    #plt.plot(x축 데이터,y축 데이터)
    plt.plot(fields, result)
    plt.show()
```

```python
townPopulation("삼성동")
```
